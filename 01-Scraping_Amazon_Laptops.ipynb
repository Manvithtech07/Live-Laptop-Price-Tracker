### **"Live" Laptop Price Tracker - End-to-End Data Engineering Project**
### Phase 1: Web Scraping

```python
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://www.amazon.in/s?k=gaming+laptop&crid=1HA5W4LWGRFV1&sprefix=gaming+lapto%2Caps%2C636&ref=nb_sb_noss_2')
```

```python
html_doc = BeautifulSoup(driver.page_source,'html.parser')
```

```python
No_of_pages = int(html_doc.find("span",class_="s-pagination-item s-pagination-disabled").text)
print(No_of_pages)
```

```python
products = html_doc.find_all("div",{'data-component-type':"s-search-result"})
```

```python
import time
from tqdm import tqdm
```

```python
data = []
for i in tqdm(range(No_of_pages)):
    url = 'https://www.amazon.in/s?k=gaming+laptop&crid=1HA5W4LWGRFV1&sprefix=gaming+lapto%2Caps%2C636&ref=nb_sb_noss_2&page=' + str(i+1)
    driver.get(url)
    html_doc = BeautifulSoup(driver.page_source,'html.parser')
    time.sleep(3)
    for product in products:
        title_tag = product.find("h2").text.strip()
        #print(title_tag) if title_tag else "N/A"
        price_tag = product.find("span", class_="a-price-whole").text.strip().replace(",", "")
        #print(price_tag) if price_tag else "N/A"
        rating_tag = product.find("span", class_="a-icon-alt").text.split(" ")[0]
        #print(rating_tag) if rating_tag else "N/A"
        review_count = product.find("span", class_="a-size-mini puis-normal-weight-text s-underline-text").text.replace("(","").replace(")","")
        #print(review_count) if review_count else "N/A"
        a_tag = product.find("a", href=True)
        link = "https://www.amazon.in" + a_tag["href"] if a_tag else "N/A"
        #print(link)
        #print("---"*10)
        data.append({
        'Title': title_tag,
        'Price': price_tag,
        'Rating': rating_tag,
        'Reviews': review_count,
        'Link': link
        })
```

```python
df = pd.DataFrame(data)
df.to_csv("amazon_laptops",index=False)
```
