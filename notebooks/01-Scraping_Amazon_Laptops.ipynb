### **"Live" Laptop Price Tracker - End-to-End Data Engineering Project**
### Phase 1: Web Scraping

```python
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://www.amazon.in/s?k=gaming+laptop&crid=1HA5W4LWGRFV1&sprefix=gaming+lapto%2Caps%2C636&ref=nb_sb_noss_2')
```

```python
html_doc = BeautifulSoup(driver.page_source,'html.parser')
```

```python
No_of_pages = int(html_doc.find("span",class_="s-pagination-item s-pagination-disabled").text)
print(No_of_pages)
```

```python
products = html_doc.find_all("div",{'data-component-type':"s-search-result"})
```

```python
import time
from tqdm import tqdm
```

```python
data = []

for i in tqdm(range(No_of_pages)):
    url = 'https://www.amazon.in/s?k=gaming+laptop&page=' + str(i+1)
    driver.get(url)
    time.sleep(3) 
    html_doc = BeautifulSoup(driver.page_source, 'html.parser')
    products = html_doc.find_all("div", {'data-component-type': "s-search-result"})
    for product in products:
        h2 = product.find("h2")
        title_tag = h2.text.strip() if h2 else None 
        
        price_span = product.find("span", class_="a-price-whole")
        price_tag = price_span.text.replace(",", "") if price_span else None
        
        rating_span = product.find("span", class_="a-icon-alt")
        rating_tag = rating_span.text.split()[0] if rating_span else None
        
        review_span = product.find("span", class_="s-underline-text")
        review_count = (review_span.text.replace("(", "").replace(")", "").replace(",", "") if review_span else None)
        
        a_tag = product.find("a", href=True)
        link = "https://www.amazon.in" + a_tag["href"] if a_tag else None
        
        data.append({
            'Title': title_tag,
            'Price': price_tag,
            'Rating': rating_tag,
            'Reviews': review_count,
            'Link': link
        })

df = pd.DataFrame(data)
df.to_csv("amazon_laptops",index=False)
print(len(df))
```
